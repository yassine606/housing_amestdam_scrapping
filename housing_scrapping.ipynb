```python
import requests 
```

```python
url = "https://www.pararius.com/apartments/amsterdam?ac=1"
page = requests.get(url)
page.status_code
```

```python
page_content = page.text
```

```python
print(page_content)
```

```python
!pip install beautifulsoup4 --upgrade 
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page_content, "html.parser")
```

```python
lists=soup.find_all('section',{"class" : "listing-search-item" })
```

```python
lists
```

```python
# extracting all the needed information 

for list in lists : 
    title= list.find("a",{'class':'listing-search-item__link--title'}).text.split()
    location=list.find("div",{"class":"listing-search-item__location"}).text.split()
    price=list.find("div",{"class":"listing-search-item__price"}).text.split()
    area=list.find("li",{"class":"illustrated-features__item illustrated-features__item--surface-area"}).text.split()
    info=[title,location,price,area]
    print(info)
    

```

```python
#creating lists containing lists : titles, prices , areas , locations 
titles=[]
locations=[]
prices=[]
area=[]
for list in lists : 
    titles.append(list.find("a",{'class':'listing-search-item__link--title'}).text.split(','))
for list in lists : 
    locations.append(list.find("div",{"class":"listing-search-item__location"}).text.split(','))
for list in lists : 
    prices.append(list.find("div",{"class":"listing-search-item__price"}).text.split(','))
for list in lists : 
    area.append(list.find("li",{"class":"illustrated-features__item illustrated-features__item--surface-area"}).text.split(','))

print(titles)
print("\n")
print(locations)
print("\n")
print(prices)
print("\n")
print(area)


```

```python
with open('housing.csv', 'w', encoding='utf8', newline='') as f:
    thewriter = writer(f)
    header = ['Title', 'Location', 'Price', 'Area']
    thewriter.writerow(header)

    for list in lists : 
        
        title= list.find("a",{'class':'listing-search-item__link--title'}).text.split()
        location=list.find("div",{"class":"listing-search-item__location"}).text.split()
        price=list.find("div",{"class":"listing-search-item__price"}).text.split()
        area=list.find("li",{"class":"illustrated-features__item illustrated-features__item--surface-area"}).text.split()
        info=[title,location,price,area]
        thewriter.writerow(info)
        
        
```
